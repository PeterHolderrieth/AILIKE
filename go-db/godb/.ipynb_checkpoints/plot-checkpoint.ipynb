```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
real_names = {"tweets_c_500" : "Tweets #Clusters = 500", "tweets_c_50" : "Tweets #Clusters = 50", 
              "tweets_c_100" : "100", "tweets_c_250" : "Tweets #Clusters = 250", 
              "tweets_c_40" : "40", "tweets_c_80" : "40000",
              "tweets_c_200" : "200",
              "tweets_10000" : "10000", "tweets_20000" : "20000", 
              "tweets_10000_c_62" : "Tweets, 10000 entries", "tweets_2500_c_16" : "Tweets, 2500 entries", 
              "tweets_10000_c_20" : "10000", "tweets_20000_c_40" : "20000", "tweets_5000_c_10" : "5000",
              "tweets_2500_c_5" : "2500", 
              "tweets" : "No Index", "tweets_c_80" : "Unclustered Index", "clustered_tweets_c_80" : "Clustered Index"
            }

vary_n = [2, 4, 8, 32, 64, 128, 256]
for n in vary_n:
    real_names[f"tweets_c_250_"+str(n)] = f"Table #Cluster 250, Limit {n}"
    real_names[f"tweets_"+str(n)] = f"Table (Without Index) #Cluster 250, Limit {n}"
```

```python
def plot_speedup(dir, comparisons, name="result", time_file="ALL_TIMINGS.csv", savefig="speedup", x_label='table'):
    df = pd.read_csv(f"{dir}/{time_file}", header=None)
    print(df)
    speedups = []
    for table in comparisons:
        base = int(df[df[0]==comparisons[table]][1].iloc[0])
        comp = int(df[df[0]==table][1].iloc[0])
        speedups.append({'table': real_names[table], 'speedup' : base/comp})

    result = pd.DataFrame(speedups)
    font_properties = {'family': 'serif', 'weight': 'normal', 'size': 10}
    ax = result.plot(kind='bar', x='table', y='speedup', legend=False, color='none', edgecolor='black')
    for p in ax.patches:
        ax.annotate(str(int(p.get_height())) + "x", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(p.get_x(), 5), textcoords='offset points', fontproperties=font_properties)

    font_properties = {'family': 'serif', 'weight': 'normal', 'size': 14}
    # plt.title(name, fontdict=font_properties)
    plt.xlabel(x_label, fontdict=font_properties)
    plt.ylabel('Speedup', fontdict=font_properties)
    
    plt.xticks(fontproperties='serif', fontsize=14, color='black', rotation=0)
    plt.yticks(fontproperties='serif', fontsize=14, color='black')

    fig = plt.gcf()
    # Set the size of the figure in inches
    fig.set_size_inches(6, 5)

    # plt.savefig(f'{dir}/{savefig}.png')
    # plt.savefig(f'{dir}/{savefig}.pdf')
    fig.savefig(f'{dir}/{savefig}.png', bbox_inches='tight')
    fig.savefig(f'{dir}/{savefig}.pdf', bbox_inches='tight')
    plt.show()
            
```

```python
plot_speedup("/Users/manyab/AILIKE/go-db/godb/benchmark_results/var_num_cluster", 
             {"tweets_c_40" : "tweets",
              "tweets_c_80" : "tweets",
              "tweets_c_100" : "tweets",
              "tweets_c_200" : "tweets",}, 
              time_file = "times.csv",
              name = "Speedup Vs #Clusters", 
              savefig="speedup_cluster", x_label="Cluster Size")

plot_speedup("/Users/manyab/AILIKE/go-db/godb/benchmark_results/var_db_size", 
             {"tweets_2500_c_5" : "tweets_2500",
              "tweets_5000_c_10" : "tweets_5000",
              "tweets_10000_c_20" : "tweets_10000",
              "tweets_20000_c_40" : "tweets_20000",
              "tweets_c_80" : "tweets"}, 
              time_file = "times.csv",
              name = "Speedup Vs #Clusters", 
              savefig="speedup_dbsize", x_label="#Rows in Dataset")
```

```python
def plot_accuracy(dir, comparisons, name="result", x_label="Table", savefig="result"):
    
    percentages = []
    for table in comparisons:     
        files = [filename for filename in os.listdir(dir) if filename.startswith(table + "_")]
        percentage_local = []
        for file in files:
            # print(f"{dir}/{file.replace(table, comparisons[table])}.csv")
            # print(f"{dir}/{file}")
            with open(f"{dir}/{file.replace(table, comparisons[table])}", 'r') as f:
                base_lines_data = f.readlines()
            with open(f"{dir}/{file}", 'r') as f:
                table_file = f.readlines()
            intersection = set(base_lines_data) & set(table_file)
            percentage_local.append((len(intersection) / min(len(base_lines_data), len(table_file))) * 100)
        percentages.append({'table': real_names[table], 'overlap' : sum(percentage_local)/ len(percentage_local)})
    print(percentages)

    result = pd.DataFrame(percentages)
    ax = result.plot(kind='bar', x='table', y='overlap', legend=False, color='none', edgecolor='black')

    
    
    font_properties = {'family': 'serif', 'weight': 'normal', 'size': 10}
    for p in ax.patches:
        ax.annotate(str(int(p.get_height())) + "%", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(p.get_x(), 5), textcoords='offset points', fontproperties=font_properties)
        
    font_properties = {'family': 'serif', 'weight': 'normal', 'size': 14}
    # plt.title(name)
    plt.xlabel(x_label, fontdict=font_properties)
    plt.ylabel('Accuracy (%)', fontdict=font_properties)
    plt.xticks(fontproperties='serif', fontsize=14, color='black', rotation=0)
    plt.yticks(fontproperties='serif', fontsize=14, color='black')


    fig = plt.gcf()
    # Set the size of the figure in inches
    fig.set_size_inches(6, 5)

    fig.savefig(f'{dir}/{savefig}.png')
    fig.savefig(f'{dir}/{savefig}.pdf')
    
    plt.show()
```

```python
vary_probes = [2, 4, 8, 16]
for n in vary_probes:
    real_names[f"tweets_c_80_"+str(n)] = f"{n}"
    real_names[f"tweets_"+str(n)] = f"#Probes = {n}"

comps = {}
for num in vary_probes:
    comps[f"tweets_c_80_"+str(num)] = f"tweets_"+str(num)
    

plot_accuracy("./benchmark_results/var_probe", 
               comps,
               name = "Accuracy versus Constant added in #Probes", x_label="#Probes", savefig="acc_probes")
```

```python
marker_style =  {"tweets" : 'o', "clustered_tweets_c_80" : 's', "tweets_c_80" : 'D',
                "tweets_0" : 'o', "tweets_1" : "*",
                "clustered_tweets_c_80_0" : 'o', "clustered_tweets_c_80_1" : "*",
                 "tweets_c_80_0" : 'o', "tweets_c_80_1" : "*",
                }
color_style =  {"tweets" : 'red', "clustered_tweets_c_80" : 'green', "tweets_c_80" : 'blue',
               "tweets_0" : 'red', "tweets_1" : 'red', 
                "clustered_tweets_c_80_0" : 'green', "clustered_tweets_c_80_1" : 'green',
                "tweets_c_80_0" : 'blue', "tweets_c_80_1" : 'blue'}

line_style = {"tweets_c_80_0" : "--", "tweets_0" : "--",  "clustered_tweets_c_80_0" : "--"}

def plot_performance(dir, tables, to_vary, name="result", time_file="ALL_TIMINGS.csv", savefig="speedup", x_label="Limit Attribute",
                    log='false'):
    df = pd.read_csv(f"{dir}/{time_file}", header=None)
    print(df)
    perf = {}
    for table in tables:
        val = []
        
        for v in to_vary:
            # base = int(df[df[0]==][1].iloc[0])
            val.append(int(df[df[0]==f'{table}_{v}'][1].iloc[0]))
        print(val)
        perf[real_names[table]] = val
        
    perf['limits'] = to_vary

    df = pd.DataFrame(perf)

    for table in tables:
        linestyle = '-'
        if table in line_style:
            print(table)
            linestyle = line_style[table]
            print(linestyle)
        plt.plot(df['limits'], df[real_names[table]], marker=marker_style[table], 
                 linestyle=linestyle, 
                 color=color_style[table],
                 label=real_names[table])
    


    font_properties = {'family': 'serif', 'weight': 'normal', 'size': 14}
    plt.xlabel(x_label, fontdict=font_properties)
    plt.ylabel('Runtime (ms)', fontdict=font_properties)

    font_properties = {'family': 'serif', 'weight': 'normal', 'size': 10}
    
    plt.legend(loc=7, prop=font_properties)
    plt.xticks(fontproperties='serif', fontsize=14, color='black')
    plt.yticks(fontproperties='serif', fontsize=14, color='black')

    if log == 'true':
        plt.xscale('log')
    plt.yscale('log')

    
    fig = plt.gcf()
    # Set the size of the figure in inches
    fig.set_size_inches(6, 5)
    fig.savefig(f'{dir}/{savefig}.png', bbox_inches='tight')
    fig.savefig(f'{dir}/{savefig}.pdf', bbox_inches='tight')

    plt.savefig(f'{dir}/{savefig}.png', bbox_inches='tight')
    plt.savefig(f'{dir}/{savefig}.pdf', bbox_inches='tight')

    plt.show()
            
```

```python
comps = {}
for num in [1, 8, 32, 128]:
    comps[f"tweets_c_80_"+str(num)] = f"tweets_{num}"
    comps[f"clustered_tweets_c_80_"+str(num)] = f"tweets_{num}"
    real_names[f"tweets_c_80_"+str(num)] = f"S {num}"
    real_names[f"clustered_tweets_c_80_"+str(num)] = f"C {num}"

plot_performance("/Users/manyab/AILIKE/go-db/godb/benchmark_results/var_n", 
               ["tweets", "clustered_tweets_c_80", "tweets_c_80"],
               [1, 8, 64, 512, 4096, 8000, 16000, 40000],
               name = "Vary Limit #", time_file = "times.csv", savefig="lim_perf", log='true')
```
